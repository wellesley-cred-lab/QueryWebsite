Loop through sub folders (SERP_date) in html folder, generate an HTML page for each folder

```python
import os
```

Get the file names from the htmlPages directory

```python
folders = os.listdir('SERP_Collection')
folders.remove(".DS_Store")
print(folders)

identities = []
relationships = []

for folder in folders:
    files = os.listdir('SERP_Collection/'+folder)
    #print(files)
    #files.remove(".DS_Store")
    #len(files)
    files.sort()
    for f in files:    
        queryName = f[0:f.find(".")]
        #print(queryName)
        if queryName in identitiesList:
            filePath = "SERP_Collection/6-8-22/"+f
            identities.append({"filePath":filePath, "queryName":queryName})
            #print(filePath)
        if queryName in relationshipsList:
            filePath = "SERP_Collection/6-8-22/"+f
            relationships.append({"filePath":filePath, "queryName":queryName})
            #print(filePath)
    #print(identities)
    #print()
    #print(relationships)
    generateHTML(folder, identities, relationships)  ##Uncomment later

#generateIndexHTML(folders, identities, relationships)  ##Uncomment later
```
