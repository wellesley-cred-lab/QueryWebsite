```python
folders = os.listdir('SERP_Collection')
folders.remove(".DS_Store")
print(folders)

identities = []
relationships = []

##Go through all Date folders in the SERP_Collection folder
##Create a new HTML (List all queries and Filter by component) for every new Date folder
#folders is the SERP_Collection folder; folder is each Date folder (e.g. 6-9-22); files are all the individual query HTMLs (e.g. Asexual.html)
for folder in folders:
    files = os.listdir('SERP_Collection/'+folder)
    #print(files)
    #files.remove(".DS_Store")
    #len(files)
    files.sort()
    
    ##Sort the queries into the topics identity and relationship
    #f is an individual query HTML (e.g. "Asexual.html")
    for f in files:    
        queryName = f[0:f.find(".")]  #Extract just the query (e.g. asexual)
        #print(queryName)
        if queryName in identitiesList:  #identitiesList is the list of all queries from topic 1 (loaded from the identities query list JSON file)
            filePath = "SERP_Collection/6-8-22/"+f
            identities.append({"filePath":filePath, "queryName":queryName})
            #print(filePath)
        if queryName in relationshipsList:  #relationshipsList is the list of all queries from topic 2 (loaded from the relationships query list JSON file)
            filePath = "SERP_Collection/6-8-22/"+f
            relationships.append({"filePath":filePath, "queryName":queryName})
            #print(filePath)
    #print(identities)
    #print()
    #print(relationships)
    
    ##Add class (e.g. peoplealsoask, hotline) to each query
    #Load all the queries and their components on a particular date (loaded from components_6-9-22.json)
    ##**REMEMBER TO RUN serp-scraper.py FOR ANY NEW DATE SERP COLLECTION ENTRY -> AUTOMATE GENERATION OF components_date_.json
    components_path = 'SERP-Components'
    components_files = os.listdir(components_path)

    for cf in components_files:
        if not cf.startswith('.'):
            date = cf.split('_')[1].replace('.json', '')
            if date == folder:
                with open(f'{components_path}/{cf}', 'r') as inFile:
                    day_components = json.load(inFile) #day_components is the list of dictionary of all the components of a query SERP for the particular date
                    
    comp_queries_identities = {}
    comp_queries_relationships = {}

    for q in day_components:
        if q in identitiesList:
            comp_queries_identities[q] = day_components[q]
        else:
            comp_queries_relationships[q] = day_components[q]
                
    ##Generate an HTML page for each SERP collection date            
    generateHTML(folder, identities, relationships, comp_queries_relationships,
                 comp_queries_identities)  
    #generateRankHTML(folder, identities, relationships)  ##UNCOMMENT LATER

##Generate the home page HTML    
generateIndexHTML(folders, identities, relationships)  
```
