```python
def generateRankHTML(f,ra,fs,i,ac):
    
    indexTemplate = """
    <DOCTYPE html>
    <html>
    
    <head>
        <meta charset="utf-8">
        <meta name="viewport" content="width=device-width"> 
        <title>Query Website</title>
		<script src="../node_modules/jquery/dist/jquery.min.js"></script>
		<script src="app.js"></script>
		
    </head>

    <body>
            <div class="centered">
                <h1> Interactive Website for Google Search Results</h1>
                <p>Below are the ranks of the original 130 queries based on the amount of changes from the previous SERP collection date. Click on the links to explore the exact changes. </p>
                
                <table>
                    <tr>
                      <th>Rank</th>
                      <th>Score</th>
                      <th>Query and Link to Change</th> 
                      <th>Change Type and Frequency</th>
                    </tr>
                    
                    {% for query in ranks %}
                    <tr>
                        <td>{{ranks[query]['rank']}}</td>
                        <td>score: {{ranks[query]['score']}}</td>
                        {% for category in allCategories %}
                            {% if query in allCategories[category]%}
                                <td><a href="../comparator.html?param1={{query.replace(" ", "%20")}}&param2={{folder}}&param3={{folders[index+1]}}&param4={{category}}">{{query}}</a></td> 
                            {% endif %}
                        {% endfor %}
                        <td>appear: {{ranks[query]['appear']}}  move: {{ranks[query]['move']}}</td>
                    </tr>
                    {% endfor %}
                    
                    
                    
                  </table>

            

        <script src="app.js"></script>
        <noscript>You need to enable JavaScript to view the full site.</noscript>


        <script data-main="scripts/main" src="scripts/require.js"></script>
        <link rel="stylesheet" href="../main.css">
        
        <script src="app.js"></script>
        <noscript>You need to enable JavaScript to view the full site.</noscript>
        <canvas id="canvas" width="640" height="480"></canvas>
    </body>
    
    </html>
    """
    tempObj = Template(indexTemplate)
    
    #Render the template by passing as an argument the list of objects.
    result = tempObj.render(folder = f, ranks=ra, folders =fs, index=i, allCategories=ac)
    print(result)
    
    filePATH = 'ranked-html/rank_'+f+'.html'
    with open(filePATH, 'w') as outFile:
        outFile.write(result)
   
```

```python
# Place the Date folders in order
folders = os.listdir('SERP_Collection')
folders.remove(".DS_Store")
print(folders)
datesList = []
for folder in folders:
    dateComponents = folder.split("-")
    datesList.append({'month':int(dateComponents[0]), 'day':int(dateComponents[1]), 'year':int(dateComponents[2])})
datesList = sorted(datesList, key=lambda x: (x['year'], x['month'], x['day']),reverse=True)
sortedDates = []
for d in datesList:
    sortedDates.append(str(d['month'])+'-'+str(d['day'])+'-'+str(d['year']))
folders = sortedDates 
    


'''
Create a list of queries under each cateogry (e.g. queer, relationships)
# folders is the SERP_Collection folder; folder is the individual date folder (e.g. 6-8-22)
# categories is all the category folders inside SERP_Collection (e.g. queer, relationships)
# allCategories will be a dictionary whose keys are the categories and values are a list of all the queries under that category
# category is the individual category (e.g. queer)
# files is the list of queries with the .html extension under the category
# querynames will be the list of queries under the category
'''
for folder in folders:  
    categories = os.listdir('SERP_Collection/'+folder)    
    allCategories = {}    
    for category in categories:    
        if not category.startswith('.'):
            files = os.listdir('SERP_Collection/'+folder+'/'+category) 
            querynames = []    
            files.sort()
            for f in files:
                if not f.startswith('.') and f.endswith(".html"):  # ignore the .DStore
                    querynames.append(f.replace('.html', "")) # get rid of the .html extension
            allCategories[category] = querynames
print(allCategories)
```
